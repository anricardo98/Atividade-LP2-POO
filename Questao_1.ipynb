```
class Viajante():
  
  def diga_OI(self):
    pass
  
  
  
```

```
class Brasileiro(Viajante):
    
  def __init__(self):
    self.nascionalidade = "Brasileiro"
    
  def __str__(self):
    return ("A nacionalidade dele é " + self.nascionalidade)
    
  
  def diga_OI(self):
    return ("Oi")
```

```
class Espanhol(Viajante):
  
  def __init__(self):
    self.nascionalidade = "Espanhol"
    
  def __str__(self):
    return ("A nascionalidade dele é " + self.nascionalidade)
  
  def diga_OI(self):
    return ("Hola")
```

```
class Americano(Viajante):
  
  def __init__(self):
    self.nascionalidade = "Americano"
    
  def __str__(self):
    return ("A nascionalidade dele é " + self.nascionalidade)
  
  def diga_OI(self):
    return ("Hello")
```

```
class TestandoNacionalidade():
  
  def __init__(self, v1 = None, v2 = None, v3 = None):
    self.v1 = v1
    self.v2 = v2
    self.v3 = v3
    
  def __str__(self):
    return ("V1 disse: {}\nV2 disse: {} \nV3 disse: {}".format(self.v1.diga_OI(), self.v2.diga_OI(), self.v3.diga_OI()))
 
```

```
main =  TestandoNacionalidade(Americano(), Espanhol(), Brasileiro())
print (main)
```
