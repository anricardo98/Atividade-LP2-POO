```
import abc

class ContaBancaria(metaclass=abc.ABCMeta):
  
  def __init__(self, saldo, multiplicador):
    self.saldo = saldo
    self.saldo = self.saldo * multiplicador
  
  @property
  def saldo(self):
      return self.__saldo
    
  @saldo.setter
  def saldo(self, saldo):
      try:
        saldo = float(saldo) 
        
      except ValueError as error:
        print(f"Voce nao colocou um numero no saldo. O saldo será 0")
        self.__lado = 0
        
      if (saldo < 0):
        
        print(f"O valor colocado é negativo. O saldo da conta será 0")
        self.__saldo = 0
        
      else:
        self.__saldo = saldo
  
  @abc.abstractmethod
  def Depositar(self, input):
    return
    
    
  def Sacar(self, valor):
    
    if (valor <= self.saldo):
      self.saldo = self.saldo - valor
    else:
      print("Não é possivel realizar esta operacao")
    
  def Transferir(self, valor, conta):
    
    if (valor <= self.saldo):
      self.saldo = self.saldo - valor
      conta.Depositar(valor)
      
    else:
      print("Não é possivel realizar esta operacao")
    
  @abc.abstractmethod
  def calcularSaldo(self, input):
    return
  
  
```

```
class ContaCorrente(ContaBancaria):
  
  def __init__(self, usuario, valor = 0):
    self.usuario = usuario
    super().__init__(valor, 0.9)
    
  def __str__(self):
    return ("Olá, {0}. \nO saldo é: {1}".format(self.usuario, str(self.saldo)))
  
  def calcularSaldo(self):
    print ("O saldo será "+ str(self.saldo))
  
  def Depositar(self, valor):
    self.saldo += (0.9 * valor)
    
  @property
  def usuario(self):
      return self.__usuario
    
  @usuario.setter
  def usuario(self, usuario):
      from re import search
      if not usuario or search(r"\d", usuario):
          print("Nome invalido: o nome ou está vazio ou apresenta algum numero")
          self.__usuario = "Invalido"
      else:
          self.__usuario = usuario 
  
  @staticmethod
  def build():
    nome = input("Criando uma conta corrente\nDigite o nome do usuario: ")
    valor = input("Digite o valor inicial da conta: ")
    return (ContaCorrente(nome, valor))
```
