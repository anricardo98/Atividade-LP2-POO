```
import abc

class AreaCalculavel(metaclass=abc.ABCMeta):
  
  @abc.abstractmethod
  def calcularArea(self, input):
    return 
  
```

```
class Quadrado(AreaCalculavel):
  
  def __init__(self, lado):
    self.lado = lado
    
  def calcularArea(self):
    if (self.lado == "Invalido"):
      print ("Não é possivel calcular a área do quadrado, devido ao lado posto ser invalido")
    else: 
      print ("A área será: " + str(self.lado * self.lado))
    
  @property
  def lado(self):
      return self.__lado
    
  @lado.setter
  def lado(self, lado):
      try:
        lado = int(lado) 
        self.__lado = lado
      except ValueError as error:
        print(f"Voce nao colocou um numero no lado")
        self.__lado = "Invalido"
    
  @staticmethod
  def build():
    lado = input("Criando um quadrado\nDigite o valor do lado:")
    return Quadrado(lado)
```

```
class Retangulo(AreaCalculavel):
  
  def __init__(self, lado1, lado2):
    self.lado1 = lado1
    self.lado2 = lado2
    
  def calcularArea(self):
    if (self.lado1 == "Invalido" or self.lado2 == "Invalido"):
      print ("Não é possivel calcular a área do retangulo, devido a pelo menos um dos lados ser invalido")
    
    else:
      print ("A área será: " + str(self.lado2 * self.lado1))  
    
  @property
  def lado1(self):
      return self.__lado1
    
  @lado1.setter
  def lado1(self, lado1):
      try:
        lado1 = int(lado1) 
        self.__lado1 = lado1
      except ValueError as error:
        print(f"Voce nao colocou um numero no lado 1")
        self.__lado1 = "Invalido"
  
  @property
  def lado2(self):
      return self.__lado2
    
  @lado2.setter
  def lado2(self, lado2):
      try:
        lado2 = int(lado2) 
        self.__lado2 = lado2
      except ValueError as error:
        print(f"Voce nao colocou um numero no lado 2")
        self.__lado2 = "Invalido"
        
  @staticmethod
  def build():
    lado1 = input("Criando um retangulo\nDigite o valor de um lado:")
    lado2 = input("Digite o valor do outro lado: ")
    return Retangulo(lado1, lado2)
```

```
from math import pi

class Circunferencia(AreaCalculavel):
  
  def __init__(self, raio):
    self.raio = raio
    
  def calcularArea(self):
    if (self.raio == "Invalido"):
      print ("Não é possivel calcular a area, devido ao raio posto ser invalido")
    else:
      print ("A área será: " + str(self.raio * self.raio * pi))
    
  @property
  def raio(self):
      return self.__raio
    
  @raio.setter
  def raio(self, raio):
      try:
        raio = int(raio) 
        self.__raio = raio
      except ValueError as error:
        print(f"Voce nao colocou um numero no raio")
        self.__raio = "Invalido"
  
  @staticmethod
  def build():
    raio = input("Criando uma circunferencia\nDigite o valor do raio:")
    return Circunferencia(raio)
```
