```
class Cliente():
  
  def __init__(self, nome, endereco, telefone):
    self.nome = nome
    self.endereco = endereco
    self.telefone = telefone
    
  def __str__(self):
    return ('Nome = {0}\nEndereco = {1}\nTelefone = {2}'.format(str(self.nome), str(self.endereco), str(self.telefone)))
  
  @property
  def nome(self):
      return self.__nome
    
  @nome.setter
  def nome(self, nome):
      from re import search
      if not nome or search(r"\d", nome):
          print("Nome invalido: o nome ou está vazio ou apresenta algum numero")
          self.__nome = "Invalido"
      else:
          self.__nome = nome 
  
  @property
  def endereco(self):
      return self.__endereco
    
  @endereco.setter
  def endereco(self, endereco):
      self.__endereco = endereco
      
  @property
  def telefone(self):
      return self.__telefone
    
  @nome.setter
  def telefone(self, telefone):
      from re import search
      if not telefone or search(r"\D", telefone):
          print("Telefone invalido: o Telefone ou está vazio ou apresenta algo diferente de um numero")
          self.__telefone = "Invalido"
      else:
          self.__telefone = str(telefone)
          
```

```
class PessoaFisica(Cliente):
  
  def __init__(self, nome, endereco, telefone, cpf):
    super().__init__(nome, endereco, telefone)
    self.cpf = cpf
  
  def __str__(self):
    return (super().__str__() + "\nCpf: " + str(self.cpf))
  
  @property
  def cpf(self):
      return self.__cpf
  
  @cpf.setter
  def cpf(self, cpf):
      from re import match
      if not cpf or not match(r"[0-9]{3}[\.]?[0-9]{3}[\.]?[0-9]{3}[-]?[0-9]{2}", cpf):
          print("CPF invalido: o CPF ou está vazio ou apresenta algum erro")
          self.__cpf = "Invalido"
          
      else:
          self.__cpf = str(cpf)
          
  @staticmethod
  def build_Pessoa_Fisica():
    nome = input("Voce esta criando uma pessoa fisica\nDigite o nome da pessoa: ")
    endereco = input("Digite o endereco da pessoa: ")
    telefone = input("Digite o telefone da pessoa: ")
    cpf = input("Digite o cpf da pessoa: ")
   
    return (PessoaFisica(nome, endereco, telefone, cpf))
```

```
class PessoaJuridica(Cliente):
  
  def __init__(self, nome, endereco, telefone, cnpj, nomeFantasia):
    super().__init__(nome, endereco, telefone)
    self.cnpj = cnpj
    self.nomeFantasia = nomeFantasia
    
  def __str__(self):
    return (super().__str__() + "\nCNPJ = {0} \nNome Fantasia = {1}".format(self.cnpj, self.nomeFantasia))
  
  @property
  def cnpj(self):
      return self.__cnpj
  
  @cnpj.setter
  def cnpj(self, cnpj):
      from re import match
      if not cnpj or not match(r"[0-9]{2}[\.]?[0-9]{3}[\.]?[0-9]{3}[\/]?[0-9]{4}[-]?[0-9]{2}", cnpj):
          print("CNPJ invalido: o CNPJ ou está vazio ou apresenta algum erro")
          self.__cnpj = "Invalido" 
      else:
          self.__cnpj = str(cnpj)  
          
  @property
  def nomeFantasia(self):
      return self.__nomeFantasia
    
  @nomeFantasia.setter
  def nomeFantasia(self, nomeFantasia):
      self.__nomeFantasia = nomeFantasia
      
  @staticmethod
  def build_Pessoa_Juridica():
    nome = input("Voce esta criando uma pessoa Juridica\nDigite o nome da pessoa: ")
    endereco = input("Digite o endereco da pessoa: ")
    telefone = input("Digite o telefone da pessoa: ")
    cnpj = input("Digite o CNPJ da pessoa: ")
    nomeFantasia =input("Digite o nome Fantasia da empresa: ")
    
    return (PessoaJuridica(nome, endereco, telefone, cnpj, nomeFantasia))
```
